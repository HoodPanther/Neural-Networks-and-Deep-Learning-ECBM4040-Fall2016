#### Bullet 5: Observation on embedding pairs

**Step 1: **Run RNN to obtain word embeddings using code for part I.

````python
from hw4a import test_rnnslu
rnn, dic = test_rnnslu(win=7, emb_dimension=50, nepochs=60, verbose=False)
````

**Step 2: **Store words and their embeddings.

````python
idx2label = dict((k, v) for v, k in dic['labels2idx'].items())
idx2word = dict((k, v) for v, k in dic['words2idx'].items())
embeddings = rnn.emb.get_value()
print 'idx2label =\n', idx2label
print 'idx2word =\n', idx2word
print 'embeddings =\n', embeddings
print 'embeddings.shape =', embeddings.shape
````

**Step 3: **Link the words and their embeddings. Test one word 'about'.

````python
word = {}
for idx, val in idx2word.iteritems():
    word[val] = embeddings[idx]
print word['about']
````
